```python
from pymongo import MongoClient
import pandas as pd
import numpy as np

client = MongoClient("mongodb://localhost:27017")  
db = client["stock_prediction_regression"]  
collection = db["merged_stock_gnews"]  


data = list(collection.find())
df = pd.DataFrame(data)

df['date'] = pd.to_datetime(df['date'])
df = df.sort_values(['ticker', 'date'])

features = []

for ticker, group in df.groupby("ticker"):
    group = group.sort_values("date").copy()

    group['prev_close'] = group['close'].shift(1)
    group['return'] = group['close'].pct_change()

   
    group['ma_3'] = group['close'].rolling(window=3).mean()
    group['ma_7'] = group['close'].rolling(window=7).mean()

    
    group['diff_ma_3'] = group['close'] - group['ma_3']
    group['diff_ma_7'] = group['close'] - group['ma_7']

    group['volatility_3'] = group['return'].rolling(window=3).std()
    group['volatility_7'] = group['return'].rolling(window=7).std()

    group['future_close'] = group['close'].shift(-1)
    group['future_return'] = (group['future_close'] - group['close']) / group['close']

    features.append(group)


features_df = pd.concat(features)
features_df.dropna(inplace=True)  

model_df = features_df[[
    'ticker', 'date', 'close', 'avg_vader_score', 'news_count', 'return',
    'ma_3', 'ma_7', 'diff_ma_3', 'diff_ma_7', 'volatility_3', 'volatility_7',
    'future_return'
]]

print(model_df.head())

import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

feature_cols = [
    'close', 'avg_vader_score', 'news_count', 'return',
    'ma_3', 'ma_7', 'diff_ma_3', 'diff_ma_7', 'volatility_3', 'volatility_7'
]
target_col = 'future_return'


model_df = model_df.dropna(subset=feature_cols + [target_col])

model_df = model_df.sort_values(by="date")

split_index = int(0.8 * len(model_df))
train_df = model_df.iloc[:split_index]
test_df = model_df.iloc[split_index:]

X_train = train_df[feature_cols]
y_train = train_df[target_col]
X_test = test_df[feature_cols]
y_test = test_df[target_col]

model = xgb.XGBRegressor(
    n_estimators=100,
    max_depth=5,
    learning_rate=0.1,
    objective='reg:squarederror',
    random_state=42
)
model.fit(X_train, y_train)


y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f" MSE: {mse:.5f}")
print(f"R¬≤ Score: {r2:.5f}")
```

```python
from pymongo import MongoClient
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.metrics import mean_squared_error, r2_score

client = MongoClient("mongodb://localhost:27017")  
db = client["stock_prediction_regression"]  
collection = db["merged_stock_gnews"]  

data = list(collection.find())
df = pd.DataFrame(data)

df['date'] = pd.to_datetime(df['date'])
df = df.sort_values(['ticker', 'date'])

features = []

for ticker, group in df.groupby("ticker"):
    group = group.sort_values("date").copy()

    group['prev_close'] = group['close'].shift(1)
    group['return'] = group['close'].pct_change()

    group['ma_3'] = group['close'].rolling(window=3).mean()
    group['ma_7'] = group['close'].rolling(window=7).mean()

    group['diff_ma_3'] = group['close'] - group['ma_3']
    group['diff_ma_7'] = group['close'] - group['ma_7']

    group['volatility_3'] = group['return'].rolling(window=3).std()
    group['volatility_7'] = group['return'].rolling(window=7).std()

    # üîÅ Predicting 3-day ahead return instead of next-day
    group['future_close'] = group['close'].shift(-3)
    group['future_return'] = (group['future_close'] - group['close']) / group['close']

    features.append(group)

features_df = pd.concat(features)
features_df.dropna(inplace=True)

# üîÅ Filter only rows with sentiment signal
features_df = features_df[features_df['news_count'] > 0]

model_df = features_df[[
    'ticker', 'date', 'close', 'avg_vader_score', 'news_count', 'return',
    'ma_3', 'ma_7', 'diff_ma_3', 'diff_ma_7', 'volatility_3', 'volatility_7',
    'future_return'
]]

model_df = model_df.dropna(subset=model_df.columns)  # full row NA check
model_df = model_df.sort_values(by="date")

split_index = int(0.8 * len(model_df))
train_df = model_df.iloc[:split_index]
test_df = model_df.iloc[split_index:]

X_train = train_df.drop(columns=['ticker', 'date', 'future_return'])
y_train = train_df['future_return']
X_test = test_df.drop(columns=['ticker', 'date', 'future_return'])
y_test = test_df['future_return']

model = xgb.XGBRegressor(
    n_estimators=100,
    max_depth=5,
    learning_rate=0.1,
    objective='reg:squarederror',
    random_state=42
)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f" MSE: {mse:.5f}")
print(f"R¬≤ Score: {r2:.5f}")
```

```python
from pymongo import MongoClient
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.metrics import mean_squared_error, r2_score

# Connect to MongoDB
client = MongoClient("mongodb://localhost:27017")  
db = client["stock_prediction_regression"]  
collection = db["merged_stock_gnews"]  

# Load data
data = list(collection.find())
df = pd.DataFrame(data)

# Preprocess
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values(['ticker', 'date'])

features = []

for ticker, group in df.groupby("ticker"):
    group = group.sort_values("date").copy()

    group['prev_close'] = group['close'].shift(1)
    group['return'] = group['close'].pct_change()

    group['ma_3'] = group['close'].rolling(window=3).mean()
    group['ma_7'] = group['close'].rolling(window=7).mean()

    group['diff_ma_3'] = group['close'] - group['ma_3']
    group['diff_ma_7'] = group['close'] - group['ma_7']

    group['volatility_3'] = group['return'].rolling(window=3).std()
    group['volatility_7'] = group['return'].rolling(window=7).std()

    # Future prediction: 3-day ahead log return
    group['future_close'] = group['close'].shift(-3)
    group['future_return'] = np.log(group['future_close'] / group['close'])

    # Lagged sentiment and news features
    group['prev_sentiment'] = group['avg_vader_score'].shift(1)
    group['prev_news_count'] = group['news_count'].shift(1)

    # Interaction feature
    group['sentiment_x_news'] = group['avg_vader_score'] * group['news_count']

    features.append(group)

# Combine all tickers
features_df = pd.concat(features)

# Keep only rows where news was present
features_df = features_df[features_df['news_count'] > 0]

# Drop NAs
features_df.dropna(inplace=True)

# Final features
model_df = features_df[[
    'ticker', 'date', 'close', 'avg_vader_score', 'news_count',
    'prev_sentiment', 'prev_news_count', 'sentiment_x_news', 'return',
    'ma_3', 'ma_7', 'diff_ma_3', 'diff_ma_7',
    'volatility_3', 'volatility_7', 'future_return'
]]

# Train/test split
model_df = model_df.sort_values(by="date")
split_index = int(0.8 * len(model_df))
train_df = model_df.iloc[:split_index]
test_df = model_df.iloc[split_index:]

# Prepare X and y
X_train = train_df.drop(columns=['ticker', 'date', 'future_return'])
y_train = train_df['future_return']
X_test = test_df.drop(columns=['ticker', 'date', 'future_return'])
y_test = test_df['future_return']

# Train model
# model = xgb.XGBRegressor(
#     n_estimators=100,
#     max_depth=5,
#     learning_rate=0.1,
#     objective='reg:squarederror',
#     random_state=42
# )
model = xgb.XGBRegressor(
    n_estimators=300,
    max_depth=6,  # ‚¨ÜÔ∏è
    learning_rate=0.03,  # ‚¨áÔ∏è (slower but smoother learning)
    subsample=0.8,
    colsample_bytree=0.8,
    objective='reg:squarederror',
    random_state=42
)
model.fit(X_train, y_train)

# Evaluate
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f" MSE: {mse:.5f}")
print(f"R¬≤ Score: {r2:.5f}")
```
